```python
#read the data
import pandas as pd
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
star_wars.head(10)
```

```python
#Review the column names
star_wars.columns
```

```python
# remove any rows where RespondentID is NaN
star_wars = star_wars.loc[star_wars.loc[:,'RespondentID'].notnull(),:]
```

```python
star_wars.loc[:,'RespondentID'].value_counts(dropna=False)
```

```python
#define a dictionary to define mapping
yes_no = {'Yes':True,'No':False}
```

```python
star_wars.loc[:,'Have you seen any of the 6 films in the Star Wars franchise?']=star_wars.loc[:,'Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no)
star_wars.loc[:,'Do you consider yourself to be a fan of the Star Wars film franchise?']=star_wars.loc[:,'Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)
```

```python
star_wars.head()
```

```python
# Convert each column above so that it only contains the values True and False.
import numpy as np
movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_mapping)
```

```python
star_wars.rename(columns={
        "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4": "seen_2",
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5",
        "Unnamed: 8": "seen_6"
        },inplace = True)
```

```python
star_wars.head()
```

```python
#Convert each of the columns above to a float type.
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
# rename the column
for i in range(9,16):
    star_wars.rename(columns={star_wars.columns[i]:'ranking_{}'.format(i-8)},inplace = True)
```

```python
star_wars.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Use the pandas.DataFrame.mean() method to compute the mean of each of the ranking columns from the last screen.
star_wars.iloc[:,9:15].mean().plot(kind='bar')
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.bar(range(6), star_wars[star_wars.columns[9:15]].mean())
```

Rankings
So far, we've cleaned up the data, renamed several columns, and computed the average ranking of each movie. As I suspected, it looks like the "original" movies are rated much more highly than the newer ones.

```python
star_wars.loc[:,star_wars.columns[3:9]].sum()
```

```python
plt.bar(range(6),star_wars.loc[:,star_wars.columns[3:9]].sum())
plt.show()
```

<strong>View counts</strong>
It appears that the original movies were seen by more respondents than the newer movies. This reinforces what we saw in the rankings, where the earlier movies seem to be more popular.
