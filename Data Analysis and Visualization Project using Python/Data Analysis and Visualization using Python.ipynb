```python
ax1 = recent_grads.plot(x='Sample_size',y='Median',kind = 'scatter',title='Employed vs. Sample_size')
ax1.set_title('Employed vs. Sample_size')
```

```python
ax2 = recent_grads.plot(x='Sample_size',y='Unemployment_rate',kind = 'scatter',title='Employed vs. Sample_size')
ax2.set_title('Employed vs. Unemployment_rate')
```

```python
ax3 = recent_grads.plot(x='Full_time',y='Median',kind = 'scatter',title='Employed vs. Sample_size')
ax3.set_title('Full_time vs. Median')
```

```python
ax4 = recent_grads.plot(x='ShareWomen',y='Unemployment_rate',kind = 'scatter',title='Employed vs. Sample_size')
ax4.set_title('ShareWomen vs. Sample_size')
```

```python
ax5 = recent_grads.plot(x='Men',y='Median',kind = 'scatter',title='Employed vs. Sample_size')
ax5.set_title('Men vs. Median')
```

```python
ax6 = recent_grads.plot(x='Women',y='Median',kind = 'scatter',title='Employed vs. Sample_size')
ax6.set_title('Women vs. Median')
```

```python
cols = ['Sample_size','Median','Employed','Full_time','ShareWomen','Unemployment_rate','Men','Women']
fig =plt.figure(figsize = (5,12))

for i in range(0,8):
    ax = fig.add_subplot(8,1,i+1)
    ax = recent_grads.loc[:,cols[i]].plot(kind = 'hist',rot = 40)
```
